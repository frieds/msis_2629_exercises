### Import Modules

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as tick
import calendar
% matplotlib inline
```

### Flights Dataset

```python
df_flights = sns.load_dataset('flights')
```

```python
df_flights.head()
```

#### Initial Graph

```python
df_flights.groupby('year')['passengers'].sum().plot(kind='bar', rot=0);
```

### Second Iteration of Graph

```python
def format_ticks(tick_val, pos):
    if tick_val > 1000000:
        val = int(tick_val/1000000)
        return '{:}M'.format(val)
    elif tick_val >= 1000:
        val = int(tick_val/1000)
        return '{:}K'.format(val)
    elif tick_val < 1000:
        return int(tick_val)
    else:
        return tick_val
```

```python
with sns.plotting_context("poster", font_scale=1.25):
    ax = df_flights.groupby('year')['passengers'].sum().plot(kind='bar', rot=0, figsize=(19, 10))
    plt.grid(linestyle='-', linewidth=0.67)
    plt.xlabel("")
    plt.title("Count of Passengers on Flights", fontsize=38, y=1.03)
    plt.xlabel("date [YYYY]", labelpad=16)
    plt.ylabel("count [passengers]", labelpad=16)
```

### Third Iteration

```python
with sns.plotting_context("poster", font_scale=1.45):
    ax = df_flights.groupby('year')['passengers'].sum().plot(kind='area', rot=0, figsize=(19, 8), color='darkgreen')
    plt.xlabel("")
    ax = plt.gca()
    ax.yaxis.set_major_formatter(tick.FuncFormatter(format_ticks))
```

### Final Iteration

```python
with sns.plotting_context("poster", font_scale=1.45):
    ax = df_flights.groupby('year')['passengers'].sum().plot(kind='area', rot=0, figsize=(19, 8), color='darkgreen', alpha=0.45)
    plt.xlabel("") # removes x-label
    ax = plt.gca()
    ax.yaxis.set_major_formatter(tick.FuncFormatter(format_ticks)) # makes K values on y values
    ax.tick_params(color='darkgray', colors='dimgray') # makes tick locators lighter and values lighter
    plt.locator_params(axis='y', nbins=3)  # reduces y-value frequency 
    plt.locator_params(axis='x', nbins=3)  # reduces y-value frequency 
    ax.spines['top'].set_visible(False) # removes top of frame
    ax.spines['right'].set_visible(False) # removes right of frame
    ax.spines['bottom'].set_visible(False) # removes bottom frame
    ax.spines['left'].set_visible(False) # removes left frame
    plt.grid(True, linestyle='-', linewidth=0.4)  # makes light color grid
```
